```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'Name': ['Strom, Mrs. Wilhelm (Elna Matilda Persson)', 'Navratil Mr. Michel ("Louis M Hoffman")', 'Minahan, Miss. Daisy E'],
    'Age': [29, 36.5, 33],
    'Sex': ['female', 'male', 'female']
})

display(df1)
```

```python
df2 = pd.read_csv('titanic_csv.csv', sep = ';')
display(df2)
```

```python
df3 = pd.read_csv('https://gist.githubusercontent.com/zaryanezrya/8b4ef51c707cb16d5e88a44dc00a1bb2/raw/41230f49c6268e072dbf102672f670be256922ab/gistfile1.txt')
display(df3)
```

```python
df2.columns = df2.columns.str.lower()
df3.columns = df3.columns.str.lower()
df4 = pd.concat([df2,df3]).drop_duplicates()
display(df4)
```

```python
df4 = df4.set_index("passengerid")
df4 = df4.sort_values("passengerid")
display(df4)
```

```python
df4.info()
df4.describe()
```

```python
temp = df4.loc[2].copy()
df4.loc[2] = df4.iloc[0]
df4.iloc[0] = temp
display(df4)
```

```python
df4["sex"].map({'female': 'f', 'male': 'm'})
display(df4)
```

```python
groups = df4.groupby('ticket')['name'].count()
ticket_groups = groups[groups >= 6].index

df4[df4['ticket'].isin(ticket_groups)]
```

```python
uniquecabs = df2['cabin'].dropna().unique()
df4[df4['cabin'].isin(uniquecabs)]
```

```python
df4['birthyear'] = 1912 - df4['age']
display(df4)
```

```python
df4['companion'] = df4.groupby('cabin')['name'].transform('count') - 1
display(df4)
```

```python
temp = df4.loc[1].copy()
df4.loc[1] = df4.iloc[0]
df4.iloc[0] = temp
display(df4)
```

```python
df4.to_csv('output_titanic_csv.csv', index = True)
```
